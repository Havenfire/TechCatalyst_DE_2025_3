# Load Env

```python
from dotenv import load_dotenv
import os
# Load environment variables from .env file
load_dotenv(override=True)


# Get credentials from environment
aws_access_key_id = os.getenv('AWS_ACCESS_KEY_ID')
aws_secret_access_key = os.getenv('AWS_SECRET_ACCESS_KEY')
aws_region = os.getenv('AWS_DEFAULT_REGION')
print(aws_region)
```

```python
import os
print("System environment:", os.environ.get('AWS_DEFAULT_REGION'))
```

# Boto3

```python
import boto3

# Create S3 client
s3 = boto3.client(
    's3',
    aws_access_key_id=aws_access_key_id,
    aws_secret_access_key=aws_secret_access_key,
    region_name=aws_region
)

# List all buckets
response = s3.list_buckets()

print("Buckets in your AWS account:")
for bucket in response['Buckets']:
    print(f"- {bucket['Name']}")
```

```python
objects_list = s3.list_objects_v2(Bucket="techcatalyst-raw")
```

```python
for objects in objects_list["Contents"]:
    if objects.get('Key').endswith('.csv'):
        print(objects.get('Key'))
```

```python
for objects in objects_list["Contents"]:
    output =  objects.get('Key').split('.')
    if len(output) > 1:
        if output[1] == "csv":
            print(objects.get('Key'))
```
